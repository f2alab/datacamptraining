**LOAD DATA**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the dataset
data_dict = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'age': [25, 35, 30, 40, 38, 43],
    'height': [5.5, 6.0, 5.8, 5.9, 5.7, 6.1],
    'gender': ['F', 'M', 'M', 'M', 'F', 'M'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia'],
    'salary': [5000, 6000, 7000, 8000, 9000, 10000],
    'department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales', 'Operations'],
}
df = pd.DataFrame(data_dict)

df.to_csv('employees_data.csv')
df.to_excel('employees_data.xlsx')
# Display the first few rows of the dataframe
print(df.head())


```

**SUMMARY STATISTICS**

```python
# Display the summary statistics of the dataframe
print(df['salary'].describe())
```

**PLOTS**

```python

# Create a bar plot for the average salary by department
plt.figure(figsize=(5, 3))
a = sns.scatterplot(x='age', y='salary', data=df, hue='gender', palette={'F': 'red', 'M': 'blue'})
g = sns.relplot(x='age', y='salary', data=df, kind='scatter', hue='gender', palette={'F': 'red', 'M': 'blue'})
print(type(a))
print(type(g))
```

**SEABORN TIPS DATASET**

```python
tips = sns.load_dataset('tips')

# Display the first few rows of the tips dataset
print(tips.head())

# Display the summary statistics of the tips dataset
print(tips.describe())
```

```python
#Display Relplot of the tips dataset

plt.figure(figsize=(10, 3))
sns.relplot(x='total_bill',
             y='tip', 
             data=tips, 
             kind='scatter',
             size='size',
             hue='size',
             style='smoker',)
```

```python
# Display Catplot of the tips dataset
sns.set_style('whitegrid')
sns.set_context('paper', font_scale=1.2)
#sns.set_palette('RdBu')
sns.set_palette(['purple', 'blue', 'green'])
g = sns.catplot(x='time',
             y='total_bill', 
             data=tips, 
             kind='box',
             )
g.figure.suptitle('Total Bill by Time of Day', y=1.05)
g.set_xlabels('Time of Day', fontsize=12, fontweight='bold', color='blue', )
plt.xticks(rotation=45)
```

```python
sns.regplot(x='total_bill',
                y='tip', 
                data=tips,
                ci=None,
                marker='*',)
```
